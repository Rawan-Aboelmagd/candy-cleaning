# cleaning

### 4- drop all the useless columns and  rows with too much NaN , 
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN 
- rename columns for easy access

```python
column_names = {'Q1: GOING OUT?': 'going out' ,'Q2: GENDER': 'gender' , 'Q3: AGE': 'age' , 'Q4: COUNTRY' : 'country' , 'Q5: STATE, PROVINCE, COUNTY, ETC' : 'area' , 
                'Q7: JOY OTHER' : 'joy_other' , 'Q8: DESPAIR OTHER' : 'despair_other' , 'Q9: OTHER COMMENTS' : 'other_comment', 'Q10: DRESS' : 'dress' , 'Q11: DAY' : 'day'}
df.rename(columns = column_names , inplace = True)
newdf = df.drop("Internal ID" , axis = 'columns')
newdf.dropna(axis = "columns"  , thresh=1230 , inplace = True)
newdf.dropna(axis = "rows"  , thresh=60 , inplace = True  )
newdf.count()
```

### 5 - clean `going out` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
# newdf[newdf['going out'].isna()]
mask = (newdf['going out'].isna())
newdf.loc[mask , 'going out'] = 'No'
newdf.loc[mask , 'going out']
newdf['going out'].isnull().sum()
```

### 6 - clean ` gender` column 
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
# newdf['gender'].value_counts(dropna = False)
# newdf['gender'].dtype
mask_gender = newdf['gender'].isna()
newdf.loc[mask_gender , 'gender'] = "I'd rather not say"
newdf['gender'].value_counts(dropna = False)
```

### 7 - clean ` age` column 
> tip : note the type of age column , and NaN values
 - try to use `to_numeric` 
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
newdf['age'].value_counts(dropna= False)
# newdf['age'].astype('float')
mask_age = newdf['age']
newdf['age'] = pd.to_numeric(mask_age , errors = 'coerce')
impossible = newdf['age']>100 
newdf.loc[impossible , 'age'] = 100
newdf['age'].fillna(newdf['age'].mean() , inplace = True)
mask_na = newdf['age'].isna()
newdf[mask_na]
```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways 
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
newdf['country']= df['Q4: COUNTRY']
newdf['country']= newdf['country'].str.lower().str.strip(' ')
newdf['country'].value_counts()
newdf.loc[newdf['country'] == 'the united states'  , 'country'] = 'united states'
newdf.loc[newdf['country'] == 'united state'  , 'country'] = 'united states'
newdf.loc[newdf['country'] == 'u.s.a.'  , 'country'] = 'united states'
newdf['country'].value_counts()
countires = ['states of  america' , 'canada' , 'uk', 'europe' , 'france' , 'mexico' , 'japan', 'netherlands' , 'costa rica', 'greece', 'ireland', 'korea', 'australia' , 'iceland' , 'scotland' , 'denmark', 'switzerland' , 'germany' , 'taiwan' ,'china' , 'spain' ]
for country in countires:
    matches = process.extract(country , newdf['country'] , limit = newdf.shape[0])
    print(matches)
    for potential_match in matches:
        if potential_match[1]>= 80:
            newdf.loc[newdf['country'] == potential_match[0] , 'country'] = country
newdf['country'].value_counts()
```
